```python
sentence="The Eiffel Tower was built from 1887 to 1889 by Gustave Eiffel, whose company specialized in building metal frameworks and structures."
```

```python
import nltk
nltk.download('punkt')
nltk.download('words')
nltk.download('maxent_ne_chunker')
nltk.download('averaged_perceptron_tagger')
```

```python
tokens = nltk.word_tokenize(sentence)
tag_elements = nltk.pos_tag(tokens)
named_entities = nltk.ne_chunk(tag_elements)
print(named_entities)
```

```python
import matplotlib.pyplot as plt
from nltk.tree import Tree

tokens = nltk.word_tokenize(sentence)
tag_elements = nltk.pos_tag(tokens)
named_entities = nltk.ne_chunk(tag_elements)
def draw_nltk_tree(tree):
    plt.figure(figsize=(12, 8))
    tree_text = tree.pformat()
    plt.text(0.5, 0.5, tree_text, ha='center', va='center', wrap=True, fontsize=12, family='monospace')
    plt.axis('off')
    plt.show()

draw_nltk_tree(named_entities)
```
